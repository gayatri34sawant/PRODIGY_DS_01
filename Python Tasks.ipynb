```python
import pandas as pd

import matplotlib.pyplot as plt

# Read the csv file into a data frame

df = pd.read_csv(r"C:\Users\Admin\Desktop\API_SP.POP.TOTL_DS2_en_csv_v2_85\API_SP.POP.TOTL_DS2_en_csv_v2_85.csv",skiprows=4)

print(df)
```

```python
# Filter out unnecessary columns
df = df[['Country Name','2022']]
print(df)
```

```python
# drop rows with missing values
df = df.dropna()
```

```python
# sort the data frame by popullation in decsending order
df = df.sort_values(by='2022',ascending=False)
print(df)
```

```python
# take top 10 countries for visualization

top_10_countries = df.head(10)
print(top_10_countries)
```

```python
# crete bar graph
plt.figure(figsize=(10,6))
plt.bar (top_10_countries['Country Name'],top_10_countries['2022'],color='red') 
plt.title('Top 10 countries by population in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```
